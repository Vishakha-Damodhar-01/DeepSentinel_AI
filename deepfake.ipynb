```python
import os, shutil, random

source_real = r"C:\Users\Lenovo\Downloads\ff\real_vs_fake\real-vs-fake\valid\real"
source_fake = r"C:\Users\Lenovo\Downloads\ff\real_vs_fake\real-vs-fake\valid\fake"

target_real = "data_140k/real"
target_fake = "data_140k/fake"

os.makedirs(target_real, exist_ok=True)
os.makedirs(target_fake, exist_ok=True)

real_imgs = random.sample(os.listdir(source_real), 10000)
fake_imgs = random.sample(os.listdir(source_fake), 10000)

for img in real_imgs:
    shutil.copy(os.path.join(source_real, img), target_real)

for img in fake_imgs:
    shutil.copy(os.path.join(source_fake, img), target_fake)

print("140K dataset ready ✅")
```
